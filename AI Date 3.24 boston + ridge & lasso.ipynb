```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
boston
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['target'] = boston.target
```

```python
df
```

```python
# Heatmap

df_corr=df.corr()

plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
#0. EDA (optional)

sns.pairplot(df,
             vars=['LSTAT', 'RM', 'PTRATIO', 'INDUS'],
            hue ='target')
plt.show()
```

```python
#1. LSTAT, RM, PTRTIO, INDUS 만 추출

x_data=df.loc[:, ['LSTAT', 'RM', 'PTRATIO', 'INDUS']]
y_data=df.loc[:, 'target']
```

```python
x_data.head()
```

```python
#2. Standscaler()

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_data)
x_data_scaled = scaler.transform(x_data)
x_data_scaled = pd.DataFrame(data=x_data_scaled, columns=x_data.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x_data_scaled, y_data, test_size=0.25, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#3. Prediction

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print("회귀계수(기울기):", np.round(lr.coef_, 1)) #np: numpy, coef_: 피처에 대한 회귀 계수 값
print("상수항(절편):", np.round(lr.intercept_, 1))  #intercept_: 상수항(절편) 값
```

```python
y_preds = lr.predict(X_test)
print(X_train.shape, X_test.shape)
```

```python
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
#4. 실제 데이터와 예측 데이터를 산점도 표시

plt.figure(figsize=(10,5))
plt.xlabel('LSTAT')
plt.ylabel('target')
plt.scatter(X_test['LSTAT'], y_test, label='y_test')
plt.scatter(X_test['LSTAT'], y_preds, c='r', label='y_preds')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('PTRATIO')
plt.ylabel('target')
plt.scatter(X_test['PTRATIO'], y_test, label='y_test')
plt.scatter(X_test['PTRATIO'], y_preds, c='g', label='y_preds')
plt.show() 
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('INDUS')
plt.ylabel('target')
plt.scatter(X_test['INDUS'], y_test, label='y_test')
plt.scatter(X_test['INDUS'], y_preds, c='violet', label='y_preds')
plt.show() 
```

```python
#5. 산점도위에 회기그래프(절편과 기울기, optional)

plt.figure(figsize=(10,5))
plt.xlabel('LSTAT')
plt.ylabel('target')
plt.scatter(X_test['LSTAT'], y_test, label='y_test')
plt.scatter(X_test['LSTAT'], y_preds, c='r', label='y_preds')
x = X_test['LSTAT']
y = lr.coef_[0]*x+lr.intercept_
plt.plot(x, y, color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('INDUS')
plt.ylabel('target')
plt.scatter(X_test['INDUS'], y_test, label='y_test')
plt.scatter(X_test['INDUS'], y_preds, c='violet', label='y_preds')
x = X_test['INDUS']
y = lr.coef_[3]*x+lr.intercept_
plt.plot(x, y, color='black')
plt.show()
```

```python
x_data = x_data.to_numpy()
y_data = y_data.to_numpy()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=10, include_bias=False)
poly.fit(X_train)
train_poly = poly.transform(X_train)
test_poly = poly.transform(X_test)
train_poly.shape
```

```python
lr.fit(train_poly, y_train)
print(lr.score(train_poly, y_train))
```

```python
print(lr.score(test_poly, y_test))
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_scaled, y_train)
print(ridge.score(train_scaled, y_train))
```

```python
print(ridge.score(test_scaled, y_test))
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list :
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_scaled, y_train)
    train_score.append(ridge.score(train_scaled, y_train))
    test_score.append(ridge.score(test_scaled, y_test))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

```python
ridge = Ridge(alpha=100)
ridge.fit(train_scaled, y_train)
print(ridge.score(train_scaled, y_train))
print(ridge.score(test_scaled, y_test))
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(train_scaled, y_train)
print(lasso.score(train_scaled, y_train))
```

```python
print(lasso.score(test_scaled, y_test))
```

```python
train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    lasso = Lasso(alpha=alpha, max_iter=10000)
    lasso.fit(train_scaled, y_train)
    train_score.append(lasso.score(train_scaled, y_train))
    test_score.append(lasso.score(test_scaled, y_test))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```
