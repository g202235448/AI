```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
boston
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['target'] = boston.target
```

```python
df
```

```python
# Heatmap

df_corr=df.corr()

plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
#0. EDA (optional)

sns.pairplot(df,
             vars=['LSTAT', 'RM', 'PTRATIO', 'INDUS'],
            hue ='target')
plt.show()
```

```python
#1. LSTAT, RM, PTRTIO, INDUS 만 추출

features = df[['LSTAT', 'RM', 'PTRATIO', 'INDUS']]
```

```python
features
```

```python
x_data=df.loc[:, ['LSTAT', 'RM', 'PTRATIO', 'INDUS']]
y_data=df.loc[:, 'target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x_data, y_data, test_size=0.25, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#2. Standscaler()

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

plt.scatter(X_train_scaled[:,0],X_train_scaled[:,1])
```

```python
#3. Prediction

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
print("회귀계수(기울기):", np.round(lr.coef_, 1)) #np: numpy, coef_: 피처에 대한 회귀 계수 값
print("상수항(절편):", np.round(lr.intercept_, 1))  #intercept_: 상수항(절편) 값
```

```python
y_preds = lr.predict(X_test_scaled)
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
print(lr.score(X_train_scaled, y_train))
print(lr.score(X_test_scaled, y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
print("절편 값:", lr.intercept_) # y축 절편값

# 회귀 계수(coefficient) : 독립변수의 변화에 따라 종속변수에 미치는 영향력이 크기
print("회귀계수:", np.round(lr.coef_,1))
```

```python
#4. 실제 데이터와 예측 데이터를 산점도 표시

plt.figure(figsize=(10,5))
plt.xlabel('LSTAT')
plt.ylabel('target')
plt.scatter(X_test['LSTAT'], y_test, label='y_test')
plt.scatter(X_test['LSTAT'], y_preds, c='r', label='y_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()  
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('RM')
plt.ylabel('target')
plt.scatter(X_test['RM'], y_test, label='y_test')
plt.scatter(X_test['RM'], y_preds, c='y', label='y_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('PTRATIO')
plt.ylabel('target')
plt.scatter(X_test['PTRATIO'], y_test, label='y_test')
plt.scatter(X_test['PTRATIO'], y_preds, c='g', label='y_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show() 
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('INDUS')
plt.ylabel('target')
plt.scatter(X_test['INDUS'], y_test, label='y_test')
plt.scatter(X_test['INDUS'], y_preds, c='violet', label='y_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show() 
```
