```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
wine_data = wine.data
wine_target = wine.target
```

```python
df_wine = pd.DataFrame(data=wine_data, columns=wine.feature_names)
df_lables = pd.DataFrame(data=wine_target, columns = ['label'])
```

```python
wine = pd.concat([df_wine,df_lables], axis=1)
```

```python
#Frequency distribution of species"
wine_outcome = pd.crosstab(index=wine["label"],  # Make a crosstab
                              columns="count")      # Name the count column

wine_outcome
```

```python
sns.violinplot(x="label",y="color_intensity",data=wine)
plt.show()
```

```python
x = wine.drop('label', axis=1)
```

```python
x
```

```python
x = x.to_numpy()
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```
